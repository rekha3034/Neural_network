*Classification of handwritten digits using a simple neural network*

```python
#Importing required libraries:
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Loading the train and test digits dataset from Keras
(X_train, y_train), (X_test,y_test) = keras.datasets.mnist.load_data()
```

Scaling the training data
Machine learning scaling is part of data preparation as this technique brings data points that are far from each other closer in order to increase the algorithm effectiveness and speed up the Machine Learning processing. 

```python
X_train = X_train / 255
X_test = X_test / 255
```

```python
X_train[0]
```

Flattening the dataset
Flattening is a technique that is used to convert multi-dimensional arrays into a 1-D array.
Flattening your data transforms it into a format that predictive algorithms can easily digest and learn patterns from. A well-structured dataset allows your predictive models to accurately capture the relationships and dependencies between various data points, resulting in improved predictive performance.

```python
X_train.shape
```

```python
X_train_flattened = X_train.reshape(len(X_train), 28*28)
X_test_flattened = X_test.reshape(len(X_test), 28*28)
```

```python
X_train_flattened.shape
```

```python
X_test_flattened.shape
```

```python
# 2d array converted to 1d array
X_train_flattened[0]
```

Simple neural network with no hidden layers

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape = (784,), activation = 'sigmoid')
])

model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

model.fit(X_train_flattened, y_train, epochs =5)
```

```python
model.evaluate(X_test_flattened, y_test)
```

```python
#Sample Prediction:
y_predicted = model.predict(X_test_flattened)
y_predicted
```

```python
#Sample Prediction 1:
plt.matshow(X_test[0])
```

```python
# Finding the max value: np.argmax finds a maximum element from an array and returns the index of it
np.argmax(y_predicted[0])
```

```python
y_predicted_labels = [np.argmax(i) for i in y_predicted]
y_predicted_labels[:5]
```

Confusion Matrix:
A confusion matrix is a table that helps evaluate the performance of a machine learning model, especially one that uses supervised learning. It's also known as an error matrix, and in unsupervised learning it's sometimes called a matching matrix.

```python
cm = tf.math.confusion_matrix(labels= y_test, predictions= y_predicted_labels)
cm
```

```python
# To visualise confusion matrix:
import seaborn as sns
plt.figure(figsize = (8,5))
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

Using hidden layer:
A hidden layer in a neural network is a set of intermediate layers that sit between the input and output layers.
The 'hidden'  layers are not directly observable and are responsible for the 'depth' of neural networks, enabling them to process complex data representations.

```python
# Using Keras flatten layer to reshape input into 1d array
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (28,28)),
    keras.layers.Dense(100, activation = 'relu'), # 100 neurons in the hidden layer
    keras.layers.Dense(10, activation = 'sigmoid')
])

model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

model.fit(X_train, y_train, epochs =5)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_predicted = model.predict(X_test)
y_predicted_labels = [np.argmax(i) for i in y_predicted]
cm = tf.math.confusion_matrix(labels= y_test, predictions= y_predicted_labels)
```

```python
import seaborn as sns
plt.figure(figsize = (8,5))
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
